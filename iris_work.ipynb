```python
sl1=[]
sw1=[]
sl2=[]
sw2=[]
sl3=[]
sw3=[]

pl1=[]
pw1=[]
pl2=[]
pw2=[]
pl3=[]
pw3=[]


for i in range(len(d)):
   # if( td[""][i]>0) :
        if(d["Species"][i] == "Iris-setosa"):
            sl1.append(d["SepalLengthCm"][i])
            sw1.append(d["SepalWidthCm"][i])
            pl1.append(d["PetalLengthCm"][i])
            pw1.append(d["PetalWidthCm"][i])
        elif(d["Species"][i] == "Iris-virginica"):
            sl2.append(d["SepalLengthCm"][i])
            sw2.append(d["SepalWidthCm"][i])
            pl2.append(d["PetalLengthCm"][i])
            pw2.append(d["PetalWidthCm"][i])
        else:
            d["Species"][i] == "Iris-setosa"
            sl3.append(d["SepalLengthCm"][i])
            sw3.append(d["SepalWidthCm"][i])
            pl3.append(d["PetalLengthCm"][i])
            pw3.append(d["PetalWidthCm"][i])
      
```
