```python
b=["bad","scam","hate","stressed","wrong","revenge","forced","Hilarious","boon","bane","lies","fooling","shamefull","failure","against","whooping","upset","died","crime","blunders","protest","fail","fails","wrongs","issues","loot","rip","stupid","FRIGHTENED","criticising","issue","sad"]
g=["good"]
```

```python
bt=[]
gt=[]
count=0
for i in l:
    for j in range(len(i["statuses"])):
        t=i["statuses"][j]["text"]
        count+=1
        for k in b:
            if(k in t):
                bt.append(t)
                break
```

```python
print len(bt),count
```

```python
bp=float(len(bt)*100)/count
gp=float((count-len(bt))*100)/count
```

```python
print bp,gp
```

```python
d={}
for i in l:
    for j in range(len(i["statuses"])):
        t=i["statuses"][j]["text"]
        t=t.split()
        for k in t:
            if(k not in d):
                d[k]=1
            else:
                d[k]+=1
```
