```python
w = pd.read_csv("winequality-data.csv")
```

```python
va1 = []
fa1 = []
va2 = []
fa2 = []

for i in range(len(w)):
    if(w["quality"][i]>5):
       # if(d["Outcome"]==1):
        va1.append(w["fixed.acidity"][i])
        fa1.append(w["volatile.acidity"][i])
    else:
        #w["quality"][i]>5
       # if(d["Outcome"]==1):
        va2.append(w["fixed.acidity"][i])
        fa2.append(w["volatile.acidity"][i])
       
```
